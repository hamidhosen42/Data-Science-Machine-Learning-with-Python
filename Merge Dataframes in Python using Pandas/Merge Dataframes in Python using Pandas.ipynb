```python
import pandas as pd
```

```python
df1=pd.read_csv("Merge Dataframes in Python using Pandas Data Set/Dataset1.csv")
df1
```

```python
df2=pd.read_csv("Merge Dataframes in Python using Pandas Data Set/Dataset2.csv")
df2
```

### Add two data frame

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID")
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="inner")
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="outer")
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="left")
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="right")
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="right",indicator=False)
df3
```

```python
# if two data frame combined more  than one data Frame same
df3=pd.merge(df1,df2,on="CustomerID",how="right",indicator=True)
df3
```
