```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('./DataSet/shoe.csv')
df.head()
```

```python
df.plot()
```

```python
plt.scatter(df['size(cm)'],df['class(y)'])
```

```python
df['class(y)'] = df['class(y)'].replace(['Female','Male'],[0,1])
df.head()
```

```python
df.shape
```

```python
from sklearn.linear_model import LogisticRegression

logist=LogisticRegression()
logist.fit(df[['size(cm)']],df['class(y)'])
```

```python
logist.predict(df[['size(cm)']])
```

```python
np.array(df['class(y)'])
```

```python
np.array(logist.predict_proba(df[['size(cm)']]))
```

```python
0.84531926 + 0.15468074
```

```python
logist.coef_
```

```python
logist.intercept_
```

```python
#y=mx+c
y=logist.coef_*9.500+logist.intercept_
y
```

```python
logist.score(df[['size(cm)']],df['class(y)'])*100
```
