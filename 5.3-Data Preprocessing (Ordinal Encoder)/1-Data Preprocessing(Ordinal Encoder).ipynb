```python
import pandas as pd
import numpy as np  
```

```python
df=pd.read_csv("1-Data Preprocessing  (Ordinal Encoder) DataSet/shop data.csv")
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df['income'].unique()
```

```python
income_list=['low','medium','high']
```

```python
ordinal=OrdinalEncoder(categories=[income_list])
```

```python
ordinal
```

```python
ordinal_value=ordinal.fit_transform(df[["income"]])
ordinal_value
```

```python
new_income=pd.DataFrame(ordinal_value,columns=['income'])
new_income
```

```python
df.head()
```

```python
#replace old income to new_income
df.income=new_income
```

```python
df
```

### Or

```python
#Delate income column
new_df=df.drop('income',axis=1)
new_df
```

```python
df=pd.concat([new_df,new_income],axis=1)
df
```

### Lebel Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['age']=le.fit_transform(df['age'])
```

```python
df
```

```python
# if multiple catagorical
df=df.apply(le.fit_transform)
```

```python
df
```

```python
# if multiple catagorical column then loop using

for col in df.columns:
    if df[col].dtype==np.number:
        continue
    df[col]=LabelEncoder().fit_transform(df[col])

df
```
